```python
#méthode Mélani
language = []
for row in df['language_code']:
    if row == 'eng':    
        language.append(int(1))
    elif row == 'spa':   
        language.append(int(2))
    elif row == 'fre':  
        language.append(int(3))
    elif row == 'ger':  
        language.append(int(4))
    else:           
        language.append(int(5))
df['lang'] = language
```

```python
#méthode Sylvaine
df['language_id'] = 5
df.loc[df['language_code'] == 'eng', 'language_id'] = 1
df.loc[df['language_code'] == 'fre', 'language_id'] = 3
df.loc[df['language_code'] == 'spa', 'language_id'] = 2
df.loc[df['language_code'] == 'ger', 'language_id'] = 4
```

```python
X = df.loc[:,['num_pages','ratings_count','text_reviews_count','language_id']].values
y = df.category.values
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from collections import Counter
# define pipeline
over = SMOTE(sampling_strategy='all')
#under = RandomUnderSampler(sampling_strategy=0.5)
#print(over)
steps = [('o', over)]#, ('u', under)]
pipeline = Pipeline(steps=steps)
#print(type(pipeline))
pp=pd.DataFrame([pipeline],columns=None)
#print(pp)
# transform the dataset
X, y = pipeline.fit_resample(X, y)
#print(X,y)
X_dataframe=pd.DataFrame(X,columns=['num_pages','ratings_count','text_reviews_count','language_id'])
X_dataframe['index'] = X_dataframe.index
y_dataframe=pd.DataFrame(y,columns=['category'])
y_dataframe['index'] = y_dataframe.index
#print(X_dataframe)
#print(y_dataframe)
X_y=pd.merge(X_dataframe,y_dataframe,left_on='index',right_on='index')
X_y=X_y.drop(columns=['index'])
print(X_y)


# summarize the new class distribution
counter = Counter(y)
print(counter)
```

```python
X_y.groupby(['category']).describe()
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(X_y,test_size = 0.2)
```

```python
# get the values of the columns for the training data
X_train = df_train.loc[:,['num_pages','ratings_count','text_reviews_count','language_id']].values
y_train = df_train.category.values
```

```python
X_test = df_test.loc[:,['num_pages','ratings_count','text_reviews_count','language_id']].values
y_test = df_test.category.values
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_test, y_test)
y_test_predicted=clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
report =classification_report(y_true=y_test, y_pred=y_test_predicted)
print(report)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_test, y_test)
y_test_predicted2= clf.predict(X_test)
```

```python
report2 =classification_report(y_true=y_test, y_pred=y_test_predicted2)
print(report2)
```
